```python
num_episodes = 20000
```

```python
from agent import Agent
from monitor import interact
import gym
import numpy as np
from bayes_opt import BayesianOptimization

env = gym.make('Taxi-v3')
agent = Agent(epsilon=0.1, gamma=0.9)
#agent = Agent()

avg_rewards, best_avg_reward = interact(env, agent, num_episodes)
```

```python
def interact_wrapper(epsilon, gamma, epsilon_divisor):
    agent = Agent(epsilon=epsilon, gamma=gamma, epsilon_divisor = epsilon_divisor)
    avg_rewards, best_avg_reward = interact(env, agent, num_episodes)
    return best_avg_reward
```

```python
pbounds = {'epsilon': (0.01, 0.6), 'gamma': (0.5, 1.0), 'epsilon_divisor' : (1.001,100)}

optimizer = BayesianOptimization(
    f=interact_wrapper,
    pbounds=pbounds,
    random_state=47
)

optimizer.probe(
    params={'epsilon': 0.3, 'gamma': 0.9, 'epsilon_divisor': 2},
    lazy=True,
)

optimizer.maximize(
    init_points=50,
    n_iter=500
)
```
